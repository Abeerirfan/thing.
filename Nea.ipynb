```python
import random #imports random 
total = 0  #sets value to 0
total2 = 0#sets total user 2 value as 0 
user = input("What is your name ?") #asks for user's name
user2 = input ("What is your username ?") # asks for user 2 name 

def player1 (): #creates function for player 1 
    global total #declares total variable as global so it can be used anywhere in the code
    dice = input("would you like to roll a dice? (Y/N)") #asks player to roll the dice
    if dice == "y" or dice == "Y": #if player says yes, dice is rolled.
        roll = random.randint(1,6)
        print (roll) #prints roll
        total += roll #adds number rolled to total
        print(total) #prints total
    else : #if player one doesnt say yes it goes onto player 2 
        player2() 
        
def player2 (): #same thing for player 2 as was done for player 1
    global total2
    dice2 = input("would you like to roll a dice player 2 ?")
    if dice2 == "y" or dice2 == "Y":
        roll2 = random.randint (0,6)
        print (roll2)
        total2  += roll2
        print(total2)
    else :
        player1()

for x in range (5): #repeats player one throw then player two throw 5 times.
    player1()
    player2()

if total%2 == 0:#if total is even then 10 points are added 
    total += 10
else :
    total -= 5 #else (if total is odd) then 5 points are subtracted
    
if total2%2 == 0: #same goes for player2
    total2 += 10
else :
    total2 -= 5

print ("Player One your total score is " + str(total)) # outputs the total score for player one and two
print ("Player Two your total score is " + str(total2))

while total == total2 : #if scores for both are qual then players should keep on playing
    player1()
    player2()



if total > total2 : #if total is higher than total2 then prins who won.
    winner = (user + " won ! " + str(total) )
else :
    winner = (user2 + " won !" + str (total2))

print (winner)
leader = open ("LeaderBoard.txt" , "a")
leader.write (winner)
leader.close()


    
```
